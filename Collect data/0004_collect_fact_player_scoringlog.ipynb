```python
df= pd.read_csv(r"C:\Users\elian\OneDrive\Escritorio\Jugadores Hockey Proyect\CSV\IN\jack_hughes_scoring_2023.csv")
player_name= 'Hughes, Jack'
```

```python
df.head()
```

```python
df = df.rename(columns={'Unnamed: 3': 'Home', 'Unnamed: 7': 'Goal_Strength'})
```

```python
df['Date'] = df['Date'].str.replace('-', '').astype(int)
```

```python
##Lookup game number based on date
team_games = pd.read_csv(r'C:\Users\elian\OneDrive\Escritorio\Jugadores Hockey Proyect\CSV\OUT\fact_gamelog.csv')
team_games.head(3)
```

```python
df = df.merge(team_games, on=['Date'],how='left')
df.head(3)
```

```python
gs_list = []
pa_list = []
sa_list = []

for desc in df['Description']:
    # goal scorer
    temp_gs = desc.split(' ')[2:4]
    temp_gs= ','.join(reversed(temp_gs))
    gs_list.append(temp_gs)
    # primary assist
    try:
        temp_pa = desc.split(' ')[6:8]
        print(','.join(reversed(temp_pa)))
        pa_list.append(temp_pa)
    except:
        pa_list.append(None)
    # secondary assist
    try:
        temp_sa = desc.split(' ')[9:11]
        print(','.join(reversed(temp_sa)))
        sa_list.append(temp_sa)
    except:
        sa_list.append(None)
df['Goal_Scorer'] = gs_list
df['Primary_Assist'] = pa_list
df['Secondary_Assist'] = sa_list
df['tracked_player_Name'] = player_name
```
